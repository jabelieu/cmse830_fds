```python
def calculate_score(tis_scores, bvs_scores, dse_scores):
    tis_avg = sum(tis_scores) / len(tis_scores) * 0.30
    bvs_avg = sum(bvs_scores) / len(bvs_scores) * 0.35
    dse_avg = sum(dse_scores) / len(dse_scores) * 0.35
    return round(tis_avg + bvs_avg + dse_avg, 2)
```
