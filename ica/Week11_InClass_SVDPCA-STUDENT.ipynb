````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
````

````python
planets_df = sns.load_dataset('planets')
X = planets_df.dropna().select_dtypes(include=[np.number]).drop(['number', 'year'], axis=1)
feature_names = X.columns

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

### Step 2: Perform SVD
# Perform SVD
U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)
# Calculate explained variance ratio
explained_variance_ratio = (S ** 2) / (S ** 2).sum()
### Step 3: Create the Biplot
def create_biplot(X_scaled, feature_names, pc1=0, pc2=1, scale_arrows=1):
    # Perform SVD
    U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)
    # Calculate scores (projected data points)
    scores = U * S
    # Get loadings (eigenvectors)
    loadings = Vt.T
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(10, 8))
    # Plot scores
    ax.scatter(scores[:, pc1], scores[:, pc2], alpha=0.5)
    # Plot feature vectors
    for i, feature in enumerate(feature_names):
        ax.arrow(0, 0,
                loadings[i, pc1] * scale_arrows,
                loadings[i, pc2] * scale_arrows,
                color='r', alpha=0.5)
        ax.text(loadings[i, pc1] * scale_arrows * 1.15,
                loadings[i, pc2] * scale_arrows * 1.15,
                feature, color='r')
    # Add labels and title
    ax.set_xlabel(f'PC{pc1+1} ({explained_variance_ratio[pc1]:.1%} explained var.)')
    ax.set_ylabel(f'PC{pc2+1} ({explained_variance_ratio[pc2]:.1%} explained var.)')
    ax.set_title('Biplot')
    # Add grid
    ax.grid(True)
    # Make equal aspect ratio
    ax.set_aspect('equal')
    return fig, ax
# Create the biplot
for i in range ( 2 ) :

    fig, ax = create_biplot(X_scaled, feature_names,pc1=i, pc2=i+1,)
    plt.tight_layout()
    plt.show()

fig, ax = create_biplot(X_scaled, feature_names,pc1=0, pc2=2,)
plt.tight_layout()
plt.show()
````

# Guide: How To Create a Biplot
<div style="background-color: #fff3e6; padding: 20px; border-radius: 10px; margin: 10px 0;">

This guide will walk you through creating a biplot from your dataset using Python. We'll use the iris dataset as an example, but you can apply these steps to any dataset.

### Prerequisites
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

### Step 1: Prepare Your Data
First, we need to prepare our data by:
1. Separating features from target variables
2. Standardizing the features

```python
# Load data (using iris as example)
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Step 2: Perform SVD
The Singular Value Decomposition (SVD) is the foundation of our biplot:

```python
# Perform SVD
U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)

# Calculate explained variance ratio
explained_variance_ratio = (S ** 2) / (S ** 2).sum()
```

### Step 3: Create the Biplot
Now we'll create the actual biplot. We'll plot:
- The projected data points (scores)
- The feature vectors (loadings)

```python
def create_biplot(X_scaled, feature_names, pc1=0, pc2=1, scale_arrows=1):
    # Perform SVD
    U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)
    
    # Calculate scores (projected data points)
    scores = U * S
    
    # Get loadings (eigenvectors)
    loadings = Vt.T
    
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot scores
    ax.scatter(scores[:, pc1], scores[:, pc2], alpha=0.5)
    
    # Plot feature vectors
    for i, feature in enumerate(feature_names):
        ax.arrow(0, 0,
                loadings[i, pc1] * scale_arrows,
                loadings[i, pc2] * scale_arrows,
                color='r', alpha=0.5)
        ax.text(loadings[i, pc1] * scale_arrows * 1.15,
                loadings[i, pc2] * scale_arrows * 1.15,
                feature, color='r')
    
    # Add labels and title
    ax.set_xlabel(f'PC{pc1+1} ({explained_variance_ratio[pc1]:.1%} explained var.)')
    ax.set_ylabel(f'PC{pc2+1} ({explained_variance_ratio[pc2]:.1%} explained var.)')
    ax.set_title('Biplot')
    
    # Add grid
    ax.grid(True)
    
    # Make equal aspect ratio
    ax.set_aspect('equal')
    
    return fig, ax

# Create the biplot
fig, ax = create_biplot(X_scaled, feature_names)
plt.tight_layout()
plt.show()
```

### Understanding the Biplot

The biplot shows two types of information:
1. **Data points** (dots): These are your observations projected onto the principal components
2. **Arrows**: These represent your original features
   - Direction: Shows how each feature contributes to the principal components
   - Length: Indicates the importance of the feature
   - Angle between arrows: Shows correlation between features
     - Small angle  positive correlation
     - 90 angle  no correlation
     - 180 angle  negative correlation

### Tips for Interpretation

1. Look for clustering patterns in the data points
2. Observe which features point in similar directions
3. Note which features have the longest arrows (most influential)
4. Check the explained variance percentages on the axes

### Common Issues and Solutions

1. **Arrows too long/short?**
   Adjust the `scale_arrows` parameter in the `create_biplot` function
   ```python
   fig, ax = create_biplot(X_scaled, feature_names, scale_arrows=0.5)  # Scale down arrows
   ```

2. **Points too crowded?**
   Adjust the alpha value in the scatter plot
   ```python
   ax.scatter(scores[:, pc1], scores[:, pc2], alpha=0.3)  # More transparent
   ```

3. **Want to show different PCs?**
   Specify different PC indices
   ```python
   fig, ax = create_biplot(X_scaled, feature_names, pc1=1, pc2=2)  # Show PC2 vs PC3
   ```
