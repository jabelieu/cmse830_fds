## Overview

Today is all about **learning from each other** in a supportive, informal setting. You'll share your midterm projects with your group, discuss your approaches, and teach each other "tricks" that will improve everyone's work. **What you'll gain:**
- See diverse approaches to the same assignment
- Learn visualization techniques and Streamlit tricks
- Get early feedback in a friendly environment
- Discover storytelling strategies you hadn't considered
- Build a toolkit of ideas to improve your project

## Part 1: Individual Presentations (8-10 min each)

Each person takes a turn presenting their project informally. This is **conversational**, not a formal presentation.

### What to Show (5 minutes):
- Quick demo of your app - what does it do?
- Walk through 2-3 key features or visualizations
- Show something you're proud of
- Show something you're struggling with

### What to Discuss (3-5 minutes):
Using the questions below, have a **conversation** with your group. You don't need to answer all questions - pick the ones most relevant to your project or that will help you most.

### Discussion Questions

*Use these to guide your conversations - not a checklist to complete!*

#### üéØ Project Narrative & Goals

**For the presenter to discuss:**

1. **What is the main question or problem your project addresses?** Can you state it in one sentence?
2. **What story are you trying to tell with your data?** Walk us through the narrative arc of your app.
3. **Who is your intended audience?** (General public? Domain experts? Decision-makers?)
4. **What insight do you want users to leave with?**

**For the group to explore:**

5. **Is the story clear from the presenter's demo?** What questions do you still have?
6. **How would you explain this project to someone else?** (If you can't, the narrative needs work!)

#### üìä Data Collection & Preparation

**For the presenter to share:**

7. **What are your two (or more) data sources?** Where did you find them?
8. **What challenges did you face integrating multiple datasets?** How did you overcome them?
9. **What data cleaning steps were most critical?** Why?
10. **Did you discover any surprising data quality issues?** How did you handle them?

**Group discussion:**

11. **What data wrangling trick or resource would you recommend to others?**
12. **Has anyone else dealt with similar data challenges?** Share solutions!

#### üìà Visualization Strategy

**For the presenter to explain:**

13. **What are your 3+ visualization types?** Why did you choose each one?
14. **Which visualization is most effective at communicating your message?** Why?
15. **What visualization choices did you struggle with?** (Color schemes? Chart types? Layout?)
16. **How did you ensure your encodings are appropriate?** (e.g., categorical vs. continuous data)

**Group brainstorm:**

17. **What visualization technique, library, or design principle did you discover that others should try?**
18. **Does the group have suggestions for alternative visualizations?**
19. **What visualization from this presentation would you want to learn how to make?**

#### üîß Data Processing & Missing Data

**For the presenter:**

20. **How much missing data did you encounter?** In which variables?
21. **What imputation technique(s) did you use?** Why did you choose that approach?
22. **How did you validate that your imputation was reasonable?**
23. **Did you handle missingness differently for different variables?** Explain your reasoning.

**Group sharing:**

24. **Has anyone used a different imputation approach?** What worked or didn't work?

#### üíª Streamlit App Features

**For the presenter to demonstrate:**

25. **Walk us through your app structure.** How many pages/sections do you have?
26. **What are your 2+ interactive elements?** (Sliders? Dropdowns? Buttons?)
27. **What Streamlit feature was hardest to implement?** How did you figure it out?
28. **How does your app guide users through your narrative?**
29. **Is your app self-documenting?** How do new users know what to do?

## Part 2: Group Synthesis & Trick Sharing (Final 15 minutes)

After everyone has presented, come together as a group for collaborative learning.

### üéì Share Your "Tricks"

Go around the group and each person shares:

**1. Streamlit Trick (2 min per person)**
- Demonstrate or explain one Streamlit technique you figured out
- Share the code snippet if possible
- Explain where you learned it (documentation, Stack Overflow, experimentation?)
- Common examples:
  - Layout tricks (columns, expanders, tabs)
  - File uploaders or downloads
  - Interactive widgets

**2. Visualization Technique (1-2 min per person)**
- Show or describe one visualization you're proud of
- Explain the design decisions behind it
- Share any libraries or resources that helped (Plotly, Seaborn, Altair, etc.)
- What makes it effective?

**3. Storytelling Approach (1-2 min per person)**
- How are you structuring your narrative?
- How are you guiding users through the app?
- What's your strategy for making the message clear?
- Even if it's not perfect yet, what's your plan?
