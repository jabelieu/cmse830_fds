Let's see how to load a dataset from `sklearn`. Note that what is imported is a container filled with lots of different types of information. It can look like a real mess! But, we can break out each item from the container.

```python
datasets.load_digits()
```

What do you notice? You get a lot of output! This is because the `sklearn` datasets come with additional information. This can be super useful if you are not familiar with the dataset. But, you need to break things apart before you get started. Here is how you do that:

```python
# actual data
# --> examine others that might be interesting to your group
digits_data = datasets.load_digits().data
wine_data = datasets.load_wine().data
# description of the data
digits_description = datasets.load_digits().DESCR
wine_description = datasets.load_wine().DESCR
# labels for the data
digits_labels = datasets.load_digits().target_names
wine_features = datasets.load_wine().feature_names
```

```python
import numpy as np

df_planets_subset = df_planets[["number", "orbital_period", "mass", "distance"]]

# create a boolean mask: True for NaN, False for finite values
nan_mask = df_planets_subset.isna()

# convert boolean mask to integer (False becomes 0, True becomes 1)
nan_array = nan_mask.astype(int).to_numpy()

# size the plot
plt.figure(figsize=(12, 6))

# imshow with interpolation set to 'nearest' and aspect to 'auto'
im = plt.imshow(nan_array.T, interpolation='nearest', aspect='auto', cmap='viridis')

plt.xlabel('Planet Index')
plt.ylabel('Features')
plt.title('Visualizing Missing Values in a Dataset')

# y-axis tick labels to feature names
plt.yticks(range(len(df_planets_subset.columns)), df_planets_subset.columns)

# x-axis ticks
num_planets = nan_array.shape[0]
plt.xticks(np.linspace(0, num_planets-1, min(10, num_planets)).astype(int))

plt.grid(True, axis='y', linestyle='--', alpha=0.7)
```
